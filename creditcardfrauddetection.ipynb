```python
df = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
df.head()
```

```python
df.drop(['Time'], axis = 1, inplace = True)
df.drop_duplicates(inplace = True)
df.shape
```

```python
total_transactions = len(df)
regular_transactions = len(df[df.Class==0])
fraudulent_transactions = len(df[df.Class==1])
fraud_percentage = round(((fraudulent_transactions/total_transactions) * 100),3)

print(total_transactions)
print(regular_transactions)
print(fraudulent_transactions)
print(fraud_percentage)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
amount = df['Amount'].values
df['Amount'] = sc.fit_transform(amount.reshape(-1, 1))
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('Class', axis = 1).values
y= df['Class'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.60, test_size = 0.40, random_state =42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score

model = XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
print("Recall: %.2f%%" % (recall * 100.0))
```
